Import dataset from Roboflow

```python
dataset_yaml_path = f"/content/final-fire-dataset-2/data.yaml"
```

Train YOLOv11 model with 50 epochs

```python
from ultralytics import YOLO

model = YOLO("yolo11n-seg.pt")

# Train model
results = model.train(
    data=dataset_yaml_path,
    epochs=50,
    imgsz=320,
    batch=8,
    name="yolo11-fire-segmentation"
)
```

Predict and visualize the image

```python
import matplotlib.pyplot as plt
import cv2
import torch

results = model("/content/final-fire-dataset-2/test/images/oil29_jpg.rf.5099b3fc24cc6aeb3ac28d2e09622f4f.jpg")

for result in results:
    result.save(filename="prediction.jpg")

# Load and display the saved image
img = cv2.imread("prediction.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 10))
plt.imshow(img_rgb)
plt.axis("off")
plt.title("YOLOv11 Segmentation Prediction")
plt.show()
```
